<h1> Estrategia de Datos y Distribuciones </h1>

Para que el modelo sea empírico, definimos cómo se genera cada variable. 

Usaremos datos aproximados de la economía española para calibrar las medias.
<h2>1. Datos "Reales" </h2> (Variables con fuente estadística)
<li> Salario Real ($w$): Los salarios nunca siguen una distribución normal (campana perfecta), sino una Log-Normal (mucha gente cobra poco, poca gente cobra mucho).
<li>Dato Real: Salario medio España $\approx$ 25.000€. Código: np.random.lognormal(mean, sigma)
<li>Productividad de Mercado ($z$): Asumiremos que sigue una Distribución Normal.
<li>Dato: PIB por hora trabajada $\approx$ 50-60€.
<li>Gasto Público ($G_a$): Valor fijo presupuestario.
<h2>2. Hipótesis de Distribución</h2> 
(Variables sin datos exactos)Aquí es donde entra la teoría de Anisi 1111 aplicada a la falta de datos:
<li>Productividad Extramercado ($z_b$): Hipótesis de Distribución Normal, pero con una media significativamente inferior a la del mercado (ya que no tiene economías de escala).Hipótesis: Media = 10€/hora (valor de mercado de limpieza/cuidados).<li>Tecnología de Consumo ($x, l$): Al ser parámetros abstractos, usaremos una Distribución Uniforme entre límites lógicos. Esto representa la heterogeneidad total: hay gente muy eficiente consumiendo y gente que no.


Código: np.random.uniform(low, high)
